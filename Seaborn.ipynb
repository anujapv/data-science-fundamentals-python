The histogram plot shows the distribution of the data.

kde=true overlays smooth kde curve,which estimates the probability density of the data

```python
a=np.random.randint(1,50,size=100)
sns.histplot(a,bins=10) #bell inderval
```

```python
d=np.random.randint(1,50,size=100)
sns.histplot(d,kde=True,bins=10,color='m')
```

```python
data=np.random.randn(1000)  #generatimg randomdata
#kde plot
sns.kdeplot(data,fill=True, color="green")
#show the plot
plt.show()
```

#### A KDE (Kernel Density Estimate) plot in Seaborn is used to visualize the probability density function of a continuous variable.It is similar to a histogram but provides a smooth curve, making it useful for understanding the distribution of data.

# CATEGORICAL PLOT

1.BoxPlot

A boxplot (also known as a box-and-whisker plot) in Seaborn is used to visualize the distribution, spread, and outliers of a dataset.

```python
df=sns.load_dataset('penguins')
df
```

```python
sns.boxplot(x='species',y='flipper_length_mm',data=df)
plt.title('flipper length for 3 Penguin Species',size=10,color='blue')
```

```python
# to add diff colors

sns.boxplot(x='species',y='flipper_length_mm',data=df,palette='Set2')
plt.title('flipper length for 3 Penguin Species',size=10,color='blue')
```

```python
# to find the diff color palattes on seaborn

sns.color_palette()   #defauly color palette
print(sns.palettes.SEABORN_PALETTES.keys())
```

```python
sns.boxplot(x='species',y='flipper_length_mm',data=df,hue='sex')
plt.title('Flipper Length for 3 Pengin Species by Sex',size=10,color='red')
```

2.BarPlot

Barplot in seaborn is used to represent categorical data with rectangular bars

default is mean

```python
df1=sns.load_dataset('tips')
sns.barplot(x='day',y='total_bill',data=df1,estimator=np.mean,palette='plasma')
plt.title('Average Total Bill per Day',size=10,color='red')

# here if we doesnt put estimator=np.mean the output will be mean
```

```python
sns.barplot(x='sex',y='total_bill',data=df1,palette='magma')
```

3. CountPlot

A countplot in Seaborn is used to visualize the frequency distributiom of a catogorical variable

used in ml to compare

```python
sns.countplot(x='sex',data=df1,palette='viridis')
```

# MATRIX PLOT

Heat Map

it id commonly used to show the relationship btw variables,highlights patterns and detect correlation in large datasets.

correlation headtmap: shows the correlation btw diff numerical variables in dataset

```python
# annot=true : ulill value veran vendi
# cmap: heatmap nn venda clr. Greens.magma,Blues etc
# linecolor='k': linecolor blck 
#dark colr aanell highly corelated aarikkum. clr light aavunna vech corelation kjoreyum
```

```python
# load the iris dataset
data=sns.load_dataset('iris')

#select only the numeric colums for correlation
numeric_data=data.select_dtypes(include=[float,int])

#create the heatmap
sns.heatmap(numeric_data.corr(),annot=True,lw=4,cmap='Purples',linecolor='k')
```
